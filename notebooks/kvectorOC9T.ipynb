```python
from fifipy.kvector import computeDeltaVector
rootdir = '/home/dario/FIFI-LS/Calibration/OC9T/kvector/'
dvector_b105 = computeDeltaVector(rootdir+'B_105.fits',comparison=[1.6374E-01,5.8418E-07,9.0257E-01,-4.2457E-09])
dvector_b130 = computeDeltaVector(rootdir+'B_130.fits',comparison=[-4.0515E-01,5.8112E-07,1.0798E+00,-1.8704E-08])
dvector_r105 = computeDeltaVector(rootdir+'R_105.fits',comparison=[-9.6551E-01,-1.2078E-07,1.0295E+00,-4.4071E-08])
dvector_r130 = computeDeltaVector(rootdir+'R_130.fits',comparison=[-7.7907E-01,-3.4153E-07,1.2232E+00,-2.7356E-07])
```

```python
import os
rootdir = '/home/dario/FIFI-LS/Calibration/OC9T/kvector/'

deltaVectors  = [dvector_b105, dvector_b130, dvector_r105, dvector_r130]
with open(os.path.join(rootdir,'DeltaVector_Coeffs.txt'), 'w') as f:
    for DV in deltaVectors:
        f.write(DV+'\n')
        
print('# Date   ch dich   b_x         a_x         Rx_cx(U)  b_y         a_y         Rx_cy(V)')
for DV in deltaVectors:
    print(DV)
```

```python
# List of directories
from glob import glob as gb
from astropy.io import fits
import os
rootdir = '/home/dario/FIFI-LS/Calibration/OC9T/kvector/'

for i in range(30):
    filenames = '*KV{0:02d}*sw.fits'.format(i+1)
    matchfile = os.path.join(rootdir, '**', filenames)
    files = sorted(gb(matchfile, recursive=True))
    if len(files) > 0:
        with fits.open(files[0]) as hdul:
            header = hdul[0].header
            print( 'KV{0:02d}'.format(i+1), len(files), header['DICHROIC'], header['G_STRT_B'], header['KMIRRPOS'])
```
