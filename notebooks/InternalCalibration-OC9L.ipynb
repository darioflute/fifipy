# Blue D 105 - M1

```python
from glob import glob as gb
from fifipy.calib import computeAllWaves
from fifipy.io import saveSlopeFits, readData
from fifipy.fit import computeSpectra

rootdir = '/home/dario/FIFI-LS/Calibration/'
caldir = 'OC9I/flats/'
basename = 'D105M1'
for i in range(1):
    name = basename+'_'+str(i+1)
    directory = rootdir+caldir+name+'/'
    files = gb(directory+'*.fits')
    gpos, specs = computeSpectra(files)
    aor, hk, gratpos, flux = readData(files[0])
    obsdate,telpos,pos,xy,an,za,alti,wv = hk
    detchan, order, dichroic, ncycles, nodbeam, filegpid, filenum = aor
    wave, dwave = computeAllWaves(gpos, dichroic, obsdate, detchan, order)
    outname = rootdir + caldir + name + '.fits'
    saveSlopeFits(gpos, dichroic, obsdate, detchan, order, specs, wave, dwave, outname)
```

```python
from fifipy.io import readSlopeFits

rootdir = '/home/dario/FIFI-LS/Calibration/'
caldir = 'OC9I/flats/'
basename = 'D105M1'
filename = basename+'_1.fits'
g1,w1,dw1,s1 = readSlopeFits(rootdir+caldir, filename)

from fifipy.spectra import computeSplineFits, computeMedianSpatialSpectra, computeMedianSpectrum
from fifipy.io import saveMediumSpectrum
%matplotlib inline
w, spectra = computeSplineFits(w1, dw1, s1, 'B1')
spatspectra = computeMedianSpatialSpectra(spectra)
medspec = computeMedianSpectrum(w,spatspectra)
saveMediumSpectrum(w, medspec, rootdir + caldir, basename+'_BB.fits')
```

```python
from fifipy.plots import plotResponse
plotResponse('B',105,1,w,medspec,0,rootdir+caldir,'New')
```

# Comparisons with previous runs

```python
from fifipy.io import readMediumSpectrum
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family']='STIXGeneral'
rcParams['font.size']=18
rcParams['mathtext.fontset']='stix'
rcParams['legend.numpoints']=1

rootdir = '/home/dario/FIFI-LS/Calibration/'
caldirs = ['2002/CalSource/','2104/CalSource/','OC8M/flats/','OC9L/flats/']
dates = ['2/2020','4/2021','5/2021','2/2022']
styles = ['dotted', '-','--','dashdot']
signes = [1,1,1,1]
          
files = ['D105M2','D130M2','D105M1','D130M1','D105M2F1','D105R','D130R']
factors = 1./np.array([6,6,6,6,6,12,12])**2
arrays = ['B','B','B','B','B','R','R']
colors = ['gold','violet','blue','skyblue','green','lime','red']


fig,ax = plt.subplots(figsize=(16,8))
for caldir,date,style,sign in zip(caldirs,dates,styles,signes):
    for f,c,factor in zip(files,colors,factors):
        try:
            file = rootdir+caldir+f+'_BB.fits'
            w,medspec = readMediumSpectrum(file, silent=True)
            ax.plot(w,sign*medspec*factor,linestyle=style,color=c,label=f+' '+date)
        except:
            print(file, 'ratio  missing')
ax.legend()
ax.set_title('Lab flux')
ax.set_xlabel('Wavelength [$\mu$m]')
ax.set_ylabel('Flux [V/s/Hz/sqarcsec]')
plt.ylim([0, 4e-12])
plt.xlim([48,250])
plt.show()


# Print compute ratio between 4/2019 and 10/2019
caldir1 = 'OC8I/flats/'
caldir0 = 'OC9L/flats/'
sign1 = 1
sign0 = 1
fig,ax = plt.subplots(figsize=(16,8))
bratio = []
rratio = []
for f,c,a in zip(files,colors,arrays):
    try:
        file = rootdir+caldir1+f+'_BB.fits'
        w1,medspec1 = readMediumSpectrum(file, silent=True)
        file = rootdir+caldir0+f+'_BB.fits'
        w0,medspec0 = readMediumSpectrum(file, silent=True)
        medspec0 *= sign0
        medspec1 *= sign1
        medspeci = np.interp(w0,w1,medspec1)
        ax.plot(w0,medspeci/medspec0,'.',color=c, label=f)
        if a == 'B':
            idx = w0 < 105
            bratio.append(medspeci[idx]/medspec0[idx])
        else:
            rratio.append(medspeci/medspec0)
    except:
        print(file ,' is not there')
bratio = np.concatenate(bratio)
rratio = np.concatenate(rratio)
mbratio = np.nanmedian(bratio)
mrratio = np.nanmedian(rratio)
ax.axhline(mbratio, color='blue', label='Blue array')
ax.axhline(mrratio, color='red', label='Red array')
print('median blue ',mbratio)
print('median red ',mrratio)
ax.legend()
ax.set_title('Ratio Lab flux')
ax.set_xlabel('Wavelength [$\mu$m]')
ax.set_ylabel('Flux[12/21]/Flux[05/21]')
ax.set_ylim(0.8,1.4)
plt.show()


# Comparison between D105M2 and D105M2F1

files = ['D105M2','D105M2F1','D105M1']
arrays = ['B','B','B']
#colors = ['gold','violet','blue','skyblue','green','lime','red','pink']
colors = ['gold','violet','blue','green']
caldirs = ['2002/CalSource/','2104/CalSource/','OC8M/flats/','OC9L/flats/']
dates = ['2/2020','4/2021','5/2021','2/2022']
styles = ['-','dotted','--','dashdot']
signes = [1,1,1,1]

fig,ax = plt.subplots(figsize=(16,8))
for caldir,date,style,sign in zip(caldirs,dates,styles,signes):
    for f,c in zip(files,colors):
        try:
            file = rootdir+caldir+f+'_BB.fits'
            w,medspec = readMediumSpectrum(file, silent=True)
            ax.plot(w,sign*medspec,linestyle=style,color=c,label=f+' '+date)
        except:
            print(file, ' missing')
ax.legend()
ax.set_title('Lab flux')
ax.set_xlabel('Wavelength [$\mu$m]')
ax.set_ylabel('Flux [V/s/Hz/spaxel]')
#ax.set_xlim(50,80)
plt.show()

files = ['D130M2','D130M1']
arrays = ['B','B']
#colors = ['gold','violet','blue','skyblue','green','lime','red','pink']
colors = ['gold','violet','green','blue']
caldirs = ['2002/CalSource/','2104/CalSource/','OC8M/flats/','OC9L/flats/']
dates = ['2/2020','4/2021','5/2021','2/2022']
styles = ['-','dotted','--','dashdot']
signes = [1,1,1,1]

fig,ax = plt.subplots(figsize=(16,8))
for caldir,date,style,sign in zip(caldirs,dates,styles,signes):
    for f,c in zip(files,colors):
        try:
            file = rootdir+caldir+f+'_BB.fits'
            w,medspec = readMediumSpectrum(file, silent=True)
            ax.plot(w,sign*medspec,linestyle=style,color=c,label=f+' '+date)
        except:
            print(file, ' missing')
ax.legend()
ax.set_title('Lab flux')
ax.set_xlabel('Wavelength [$\mu$m]')
ax.set_ylabel('Flux [V/s/Hz/spaxel]')
#ax.set_xlim(50,80)
plt.show()

```

```python
from fifipy.io import readMediumSpectrum
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams
from scipy.interpolate import interp1d
from fifipy.io import saveMediumSpectrum
c = 299792458.e-6 # um/s

rcParams['font.family']='STIXGeneral'
rcParams['font.size']=18
rcParams['mathtext.fontset']='stix'
rcParams['legend.numpoints']=1


rootdir = '/home/dario/FIFI-LS/Calibration/'
caldirs = ['2002/CalSource/','2104/CalSource/','OC8M/flats/','OC9I/flats/']
dates = ['2/2020','4/2021','5/2021','12/2021']
styles = [ '--','-',':','dashdot']


files = ['M2','M1','R']
colors = ['blue','lime','red']
fig,ax = plt.subplots(figsize=(16,8))
for caldir, style, date in zip(caldirs, styles, dates):
    for f,col in zip(files,colors):
        try:
            file = rootdir+caldir+'D105'+f+'_BB.fits'
            w105,medspec105 = readMediumSpectrum(file, silent=True)
            file = rootdir+caldir+'D130'+f+'_BB.fits'
            w130,medspec130 = readMediumSpectrum(file, silent=True)
            int130 = interp1d(w105, medspec105 ,fill_value='extrapolate')
            ax.plot(w130,int130(w130)/medspec130,linestyle=style,color=col,label=f+' '+date)
            saveMediumSpectrum(w130, medspec130/int130(w130), rootdir + caldir, 'D130onD105_'+f+'.fits')
        except:
            print(file, ' missing')
            pass
    
ax.legend()
ax.set_xlabel('Wavelength [$\mu$m]')
ax.set_ylabel('Flux [D105/D130]')
ax.set_ylim([0,4.5])
ax.set_xlim(45,250)
ax.grid()
ax.set_title('Dichroic 105 / 130')
plt.show()
```
